## Modeling Time
### Before we start a basic model, I want to make sure we narrow down qualifications.  The first qualification I want to use is that no data would be used if the player did not play at LEAST half of the season.

```python
df = df[df['Games'] > 40]
```

### This eliminates 12 rows of data from our pool.

```python
df
```

### This gets rid of an identifier that we didn't have a place for

```python
df.drop("identifier2", axis=1, inplace=True)
```

### PER claims that the league average score is 15.  These are not average scores, so let's see only the above average scores.

```python
df = df[df["PER"] > 15]
```

```python
df
```

### Now we have 196 rows of data to use across 46 different data categories

### Some of the Percentage stats are in percentages and not numbers (i.e 25% instead of .25).  Let's change that

```python
df['Assist %'] = df['Assist %'].apply(lambda x: x/100)
```

```python
df['Block %'] = df['Block %'].apply(lambda x: x/100)
```

```python
df['Turnover %'] = df['Turnover %'].apply(lambda x: x/100)
```

```python
df['Usage %'] = df['Usage %'].apply(lambda x: x/100)
```
