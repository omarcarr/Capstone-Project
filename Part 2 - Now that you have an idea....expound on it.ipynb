#### Data Scraping
So I noticed during the initial data scraping that I could get the Per Game stats, but not the Advanced stats that I was hoping to pull from

I found a hint from ```https://stackoverflow.com/questions/46305314/using-beautifulsoup-to-scrape-tables-within-comment-tags``` that found a way of pulling the comments, so I ***improvised*** with this and found a way to pull all of the data off the page.

````python
cap = []

table = soup.find('table', {'id': 'per_game'})
for row in table.find_all('tr')[1:21]:
        skyhook = {}
        skyhook['Age'] = row.find('td', {'data-stat': 'age'}).text
        skyhook['Games'] = row.find('td', {'data-stat': 'g'}).text
        skyhook['Minutes Per Game'] = row.find('td', {'data-stat': 'mp_per_g'}).text
        skyhook['FG Per Game'] = row.find('td', {'data-stat': 'fg_per_g'}).text
        skyhook['FG Attempts Per Game'] = row.find('td', {'data-stat': 'fga_per_g'}).text
        skyhook['FG %'] = row.find('td', {'data-stat': 'fg_pct'}).text
        skyhook['3PT Per Game'] = row.find('td', {'data-stat': 'fg3_per_g'}).text
        skyhook['3PT Attempts Per Game'] = row.find('td', {'data-stat': 'fg3a_per_g'}).text
        skyhook['3PT %'] = row.find('td', {'data-stat': 'fg3_pct'}).text
        skyhook['2PT Per Game'] = row.find('td', {'data-stat': 'fg2_per_g'}).text
        skyhook['2PT Attempts Per Game'] = row.find('td', {'data-stat': 'fg2a_per_g'}).text
        skyhook['2PT %'] = row.find('td', {'data-stat': 'fg2_pct'}).text
        skyhook['Effective FG%'] = row.find('td', {'data-stat': 'efg_pct'}).text
        skyhook['FT Per Game'] = row.find('td', {'data-stat': 'ft_per_g'}).text
        skyhook['FT Attempts Per Game'] = row.find('td', {'data-stat': 'fta_per_g'}).text
        skyhook['FT %'] = row.find('td', {'data-stat': 'ft_pct'}).text
        skyhook['Offensive Reb Per Game'] = row.find('td', {'data-stat': 'orb_per_g'}).text
        skyhook['Defensive Reb Per Game'] = row.find('td', {'data-stat': 'drb_per_g'}).text
        skyhook['Total Reb Per Game'] = row.find('td', {'data-stat': 'trb_per_g'}).text
        skyhook['Assists Per Game'] = row.find('td', {'data-stat': 'ast_per_g'}).text
        skyhook['Steals Per Game'] = row.find('td', {'data-stat': 'stl_per_g'}).text
        skyhook['Blocks Per Game'] = row.find('td', {'data-stat': 'blk_per_g'}).text
        skyhook['Turnovers Per Game'] = row.find('td', {'data-stat': 'tov_per_g'}).text
        skyhook['Fouls Per Game'] = row.find('td', {'data-stat': 'pf_per_g'}).text
        skyhook['Points Per Game'] = row.find('td', {'data-stat': 'pts_per_g'}).text
        cap.append(skyhook)
````

````python
per_game = ['Age', 'Games', 'Minutes Per Game', 'FG Per Game', 'FG Attempts Per Game', 'FG %', '3PT Per Game', '3PT Attempts Per Game', '3PT %', '2PT Per Game', '2PT Attempts Per Game', '2PT %', 'Effective FG%', 'FT Per Game', 'FT Attempts Per Game', 'FT %', 'Offensive Reb Per Game', 'Defensive Reb Per Game', 'Total Reb Per Game', 'Assists Per Game', 'Steals Per Game', 'Blocks Per Game', 'Turnovers Per Game', 'Fouls Per Game', 'Points Per Game']
````

#### Advanced Stats

````python
driver = webdriver.Chrome()
driver.get('https://www.basketball-reference.com/players/a/abdulka01.html')
soup = BeautifulSoup(driver.page_source, 'lxml')
driver.quit()
table = soup.find_all('table')[14] #advanced stats
#table = soup.find_all('table')[8] #per_game stats
````

````python
adv_cap = []

adv_cap_table = soup.find('table', {'id': 'advanced'})
````

````python
adv_cap_table
````

````python
for row in adv_cap_table.find_all('tr')[1:]:
    adv_skyhook = {}
    adv_skyhook['PER'] = row.find('td', {'data-stat': 'per'}).text
    adv_skyhook['True Shooting'] = row.find('td', {'data-stat': 'ts_pct'}).text
    adv_skyhook['Total Rebounding %'] = row.find('td', {'data-stat': 'trb_pct'}).text
    adv_skyhook['Assist %'] = row.find('td', {'data-stat': 'ast_pct'}).text
    adv_skyhook['Steal %'] = row.find('td', {'data-stat': 'stl_pct'}).text
    adv_skyhook['Block %'] = row.find('td', {'data-stat': 'blk_pct'}).text
    adv_skyhook['Turnover %'] = row.find('td', {'data-stat': 'tov_pct'}).text
    adv_skyhook['Usage %'] = row.find('td', {'data-stat': 'usg_pct'}).text
    adv_skyhook['Win Shares'] = row.find('td', {'data-stat': 'ws'}).text
    adv_skyhook['Offensive Box +/-'] = row.find('td', {'data-stat': 'obpm'}).text
    adv_skyhook['Defensive Box +/-'] = row.find('td', {'data-stat': 'dbpm'}).text
    adv_skyhook['Box +/-'] = row.find('td', {'data-stat': 'bpm'}).text
    adv_skyhook['VORP'] = row.find('td', {'data-stat': 'vorp'}).text
    adv_cap.append(adv_skyhook)
````

````python
adv_cap
````

````python
advanced = ['PER', 'True Shooting', 'Total REB %', 'Assist %', 'Steal %', 'Block %', 'TOV %', 'Usage %', 'Win Shares', 'Offensive +/-', 'Defensive +/-', 'Box +/-', 'VORP']
````

````python
kareem_adv = pd.DataFrame(adv_cap, columns=advanced)
````
